```python
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
from astropy.visualization import make_lupton_rgb
```

```python
M = fits.open('/u/e/jb2854/Mergers_Project/DF_dir/TNG_FITS_dir/940000001.fits')
```

```python
M
```

```python
data = M[3].data
data_g = data[1]
data_r = data[2]
data_i = data[3]
```

```python
data_g -= np.min(data_g)
data_r -= np.min(data_r)
data_i -= np.min(data_i)

# scale by the blue image
#
```

```python
plt.figure(figsize=(8,8))
image = make_lupton_rgb(data_g, data_r, data_i, Q=10,stretch=0.5,minimum=0)
plt.imshow(image,aspect='auto')
#plt.imshow(data[0], cmap ='gist_gray_r')
```
