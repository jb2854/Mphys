```python
def load_results(run_ids):

    results = pd.read_parquet('/home/AstroPhysics-Shared/PROJECTS/MPhys_Villforth/DATA/CNN_dir/CNN_run_%05d/' %(run_id) + 'full_subhalo_results.parquet') 
    return results


#results_list, run_info = load_results([115,117, 111, 128, 132, 134])
#results_list, run_info = load_results([175,176, 177, 178])

#results_list, run_info = load_results([179, 180, 181, 182, 183, 184, 185, 186, 187, 188, 189, 115])
#results_list, run_info = load_results([181, 204, 206, 207, 208, 209, 203, 202, 201, 200, 199, 193, 178, 176, 175])
```

```python
for run_id in cnn_run_table.RUN_ID:

    opt_th = cnn_run_table.loc[cnn_run_table.RUN_ID == run_id]['TH_OPT'].values[0]
    opt_th_F1 = cnn_run_table.loc[cnn_run_table.RUN_ID == run_id]['TH_OPT_F1'].values[0]
    
    try:
        results = load_results(run_id)

        correct = results[((results.LABEL == 0) & (results.PRED_LABEL < opt_th)) | ((results.LABEL == 1) & (results.PRED_LABEL > opt_th))]
        correct_F1 = results[((results.LABEL == 0) & (results.PRED_LABEL < opt_th_F1)) | ((results.LABEL == 1) & (results.PRED_LABEL > opt_th_F1))]
        correct_05 = results[((results.LABEL == 0) & (results.PRED_LABEL < 0.5)) | ((results.LABEL == 1) & (results.PRED_LABEL > 0.5))]

        accuracy = len(correct)/len(results)
        accuracy_F1 = len(correct_F1)/len(results)
        accuracy_05 = len(correct_05)/len(results)

        print(accuracy, accuracy_F1, accuracy_05)

        cnn_run_table.loc[cnn_run_table['RUN_ID'] == run_id, 'accuracy'] = accuracy
        cnn_run_table.loc[cnn_run_table['RUN_ID'] == run_id, 'accuracy_F1'] = accuracy_F1
        cnn_run_table.loc[cnn_run_table['RUN_ID'] == run_id, 'accuracy_05'] = accuracy_05
        
        current_time = datetime.datetime.now()

        # Use the strftime method to format the output
        formatted_time = current_time.strftime("%Y-%m-%d %H:%M:%S")

        # Print the time
        print(formatted_time)

        cnn_run_table.loc[cnn_run_table['RUN_ID'] == run_id, 'Time_stamp'] = ''


    except:
        print('Problem with run ', run_id)
```
