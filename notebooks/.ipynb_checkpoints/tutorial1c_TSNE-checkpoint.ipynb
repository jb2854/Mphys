```python
# ############## AAS 237 MACHINE LEARNING WORKSHOP, JANUARY 2021 ###############
#
# TUTORIAL 1c: t-SNE Classification
#
# GOAL: The goal of this tutorial is to learn how (unsupervised) t-Distributed Stochastic Neighbor Embedding (t-SNE)
# can be used to visualize datasets. More information on t-SNE can be found here:
# https://lvdmaaten.github.io/tsne/
# 
#
# DATA: We use the same dataset as in the previous tutorial 1a. These are Hubble H-band images from
# the CANDELS survey on the GOODS-S field. We provide ~3000 image cutouts of galaxies at low redshifts,
# which have been classified visually by Kartaltepe et al. (2015, https://ui.adsabs.harvard.edu/abs/2015ApJS..221...11K/abstract)
# in spheroidal and disk morphology.
# 
# 
# OUTLINE: We use the (unsupervised) t-SNE algorithm to group the Hubble H-band images of ~3000 galaxies.
# The algorithm is unsupervised, therefore it will group the galaxies without a training sample and project them
# onto a 2-dimensional space. We will subsequently map the galaxies back into this parameter space to understand
# the grouping performed by t-SNE, and we will find a good separation between spheroids and disks. We will also
# match the galaxies to the CANDELS 3-D HST catalog containing physical parameters of the galaxies to see how
# these variables correlate with the t-SNE grouping.
#
#
# CONTACT: Andreas Faisst (afaisst@ipac.caltech.edu)
#
#
# ####################################################
```

```python
# ========================== 1. LOAD PACKAGES ==========================
# We first load all the packages that are needed in this tutorial.
#

## a) general packages
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import pickle
import time
import random
from astropy.table import Table
from mpl_toolkits.axes_grid.inset_locator import (inset_axes)
from regions.core import PixCoord
from regions.shapes.circle import CirclePixelRegion

## b) t-SNE package from SKLEARN
from sklearn import manifold

## c) Finally, we set up some parameters to make nice plots
mpl.rcParams['xtick.direction'] = 'in'
mpl.rcParams['ytick.direction'] = 'in'
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['xtick.top'] = True
mpl.rcParams['ytick.right'] = True
mpl.rcParams['hatch.linewidth'] = 1.5
def_cols = plt.rcParams['axes.prop_cycle'].by_key()['color']
```

```python
# ========================== 2. LOAD FILES ==========================
# Now, we load all the files (images, labels, info). The data match line by line.
# The dataset that we are going to use is in the form of pickle files.
# For more information on using pickle files, see https://docs.python.org/3/library/pickle.html 


## a) Load F160W info
## This will load a 2-dimensional array containing additional info for each galaxy (such as RA/DEC coordiantes)
with open("../data/f160w_data_info_unique.pickle","rb") as f:
    f160w_info = pickle.load(f)

## b) Load F160W images
## This will load a 3-dimensional array where the first dimension is the number of cutouts
## and the second and third dimensions are the dimensions of the cutout themselves (41x41 pixels)
with open("../data/f160w_images_unique.pickle","rb") as f:
    f160w_images = pickle.load(f)


## c) Load F160W labels
## This will load a 2-dimensional array and contains one label (1=disk, 0=spheroid) for each cutout
with open("../data/f160w_labels_unique.pickle","rb") as f:
    f160w_labels = pickle.load(f)


## d) Let's print some things for information
print("Shape of image array: ", f160w_images.shape)
print("Shape of label array: " , f160w_labels.shape)
print("Number of Spheriod: %g" % len(np.where(f160w_labels == 0)[0]) )
print("Number of Disks: %g" % len(np.where(f160w_labels == 1)[0]) )
```

```python
# ========================== 3. CREATE FLATTENED IMAGES SO t-SNE CAN READ THEM ==========================
# t-SNE needs a 2-dimensional array where the objects to classify are the rows and
# the different estimator are the columns. This is different compare to the convolutional networks
# that work in n-dimensions. Since we are dealing here with 41x41 pixel images, we have have
# to flatten them into a 41*41 array and store them in a 2d array.

## a) How many cutouts to load?
## t-SNE is running very fast, so we can load here all the images. But for testing, you could load less to make it faster.
n_load = f160w_images.shape[0]
#n_load = 1000


## b) create a 2-dimensional array with the flattened cutouts
f160w_images_tsne = np.zeros((n_load, f160w_images.shape[1]*f160w_images.shape[2] ) )
for ii in range(n_load):
    f160w_images_tsne[ii,:] = f160w_images[ii].flatten()

## c) Print the shape of the new array (2-dimensional).
## The first dimension should still be the same, the second dimension should be 41*41=1681 because we "flattened" the images.
print(f160w_images_tsne.shape)
```

```python
# ========================== 4. PLOT SOME GALAXIES ==========================
# Let's plot some of the galaxies with the corresponding labels.
# Remember that the labels are defined as 1=Disk and 0=Spheroid.
# Below, we load the first 25 cutouts.


fig = plt.figure(figsize=(15,15))

for ii in range(25):
    ax = fig.add_subplot(5,5,ii+1)
    gal_number = ii
    ax.imshow(f160w_images_tsne[gal_number].reshape(41,41))
    ax.text(0.1,0.9,["Spheroid","Disk"][f160w_labels[gal_number]],color="white",fontsize=14,va="top",ha="left",transform=ax.transAxes)

#plt.savefig("tutorial1c_fig1.pdf",bbox_inches="tight")
plt.show()
```

```python
# ========================== 5. RUN TSNE AND MAKE NICE OUTPUT TABLE ==========================
# We now can run t-SNE on the flattened image arrays.

## a) Create a t-SNE model
## We have set the different arguments to reasonable values. Surely there is room to improve the performance,
## however, after playing around with different values it seems like the results do not change much.
tsne_model = manifold.TSNE(n_components=2,perplexity=10,learning_rate=200,n_iter=500, init='pca', random_state=1)

## The initial guess of the t-SNE can be choosen by the "init" parameter. In the current example, we set it to "pca" so the algorithm
## first runs a PCA as an initial guess. You can also choose "random", in which case the t-SNE is initiallized randomly and will look
## different every time you run it (in that case you can use "random_state" to give it the same random seed if you want it to look the same).
## Note that the different random implementation make the t-SNE map look differently, but quantitatively the maps is the exactly the same.
## The "perplexity" keyword is related to the number of nearest neighbors that is used in other manifold learning algorithms. Larger datasets
## usually require a larger perplexity. By playing around with the perplexity you can make the map look nicer but the results are in the
## end very insensitive to the choice of this parameter.
## The "n_components" argument defines the number of components or dimensions. Remember that t-SNE is a dimensionality reduction algorithm,
## similar to the PCA. Usually, you want to reduce to 2 dimensions and there is no advantage to reducing it to 3 dimensions usually.

## b) "Train" the t-SNE model
## This can take a while, but t-SNE is usuall very fast (will run in ~30-40 seconds on a standard Macbook Pro)
## We also print the ellapsed time for information.
## Important: Note that t-SNE is an unsupervised Machine Learning algorithm. The labels are not used in this step, but we will
## use them later to understand the distribution of objects (here cutouts) in 2-dimensional t-SNE space.
print("Running T-SNE . . .")
t0 = time.time()
tsne_output = tsne_model.fit_transform(f160w_images_tsne)
print("Time to run TSNE: %.2fs" % (time.time() - t0))

## c) Create a nice table with the results
## The t-SNE algorithm outputs the position of each object (here cutout) in a 2-dimensional space.
## Here, we just create a nice table containing the labels (not used because unsupervised!) and position in 2-dimensional t-SNE space.
tsne_results = Table(names=["id","label","X","Y"],dtype=["int","int","f","f"])
for ii in range(tsne_output.shape[0]):
    tsne_results.add_row([ii , f160w_labels[ii], tsne_output[ii][0], tsne_output[ii][1] ])
```

```python
# ========================== 6. PLOT t-SNE SPACE WITH LABELS ==========================
# Finally, let's plot the 2=dimensional t-SNE space and color the points with our labels.
# Again, note that t-SNE arranged the objects (cutouts) in this 2-dimensional space without the labels or any training set.
# We see from the plot that we create that the t-SNE has efficiently separated disk from spheroid morphology, with a few exceptions.

## create a new figure
fig = plt.figure(figsize=(10,8))
ax1 = fig.add_subplot(1,1,1)

## plot disks (label = 1)
sel = np.where(tsne_results["label"] == 1)[0]
ax1.plot(tsne_results["X"][sel],tsne_results["Y"][sel],
         "o",markersize=2,markeredgecolor=def_cols[0],alpha=0.8,
         markeredgewidth=1,
         label="Disks")

## plot spheroids (label = 0)
sel = np.where(tsne_results["label"] == 0)[0]
ax1.plot(tsne_results["X"][sel],tsne_results["Y"][sel],
         "s",markersize=5,fillstyle='none',markeredgecolor=def_cols[1],alpha=0.8,
         markeredgewidth=1,
         label="Spheriods")

## create the axes
ax1.set_xlabel("T-SNE dimension 1" , fontsize=14)
ax1.set_ylabel("T-SNE dimension 2",fontsize=14)
ax1.legend(loc="best",fontsize=14)
ax1.tick_params(axis='both', which='major',length=5, labelsize=12,pad=10)
ax1.tick_params(axis='both',which='minor',length=3)
ax1.minorticks_on()


#plt.savefig("tutorial1c_fig2.pdf",bbox_inches="tight")
plt.show()
```

```python
# ========================== 7. VISUALIZE RESULTS (PART 1) ==========================
# The good thing about unsupervised ML techniques is that they don't need a previous classification. Therefore, we can
# use these techniques in an exploratory way.
# Assume now that we do not have a visual classification of the galaxies. The first thing after running
# the t-SNE algorithm would be to investigate the differences between galaxies at different locations
# in 2-dimensional t-SNE space. An obvious thing to do is to look at their images.
# In the following, we bin the t-SNE space and stack the images of the galaxies in each of the bins.
# Obviously, we should see differences in galaxy morphology across the t-SNE space (we know this from the previous
# figure where we painted the labels onto the points).


## Create a grid and stack the cutouts.
gridsize = 10
x_grid = np.arange(-70,90,gridsize)
y_grid = np.arange(-60,80,gridsize)
grid = [(xx,yy) for xx in x_grid for yy in y_grid]

## Create mask that will be used for normalizing image
center = PixCoord( f160w_images.shape[1]//2 , f160w_images.shape[2]//2 )
reg = CirclePixelRegion(center , radius = 10) # cutout size is 41x41px, let's choose radius 1/4 of that.
mask = reg.to_mask(mode="center") # create mask
mask_img = mask.to_image( (f160w_images.shape[1],f160w_images.shape[2]) ) # create an image mask (binary: 1=data, 0=masked)

## Stack the galaxy images in each bin defined by this grid
STACKS = []
for ii in range(len(grid)):
    
        # select all galaxies in the box
        sel = np.where( (np.abs(tsne_results["X"] - grid[ii][0]) < gridsize/2) & (np.abs(tsne_results["Y"] - grid[ii][1]) < gridsize/2) )[0]
        
        if len(sel) > 1:
            stack_flatten = np.zeros(f160w_images.shape[1]*f160w_images.shape[2])
            for ss in sel:
                
                # normalize image first (normalize max = 1)
                # We measure the maximum flux value in the center of the cutout to avoid confusion with merging galaxies
                # We use the mask that we have defined above
                img_max = np.nanmax(f160w_images[ss] * mask_img) # compute maximum of masked image
                if img_max == 0:
                    img_max = 1
                stack_flatten += f160w_images[ss].flatten() / img_max
            stack_flatten = stack_flatten / len(sel)
            STACKS.append(stack_flatten)
        else:
            STACKS.append([-99])
            

## Now create a figure similar to the one already done above
fig = plt.figure(figsize=(10,8))
ax1 = fig.add_subplot(1,1,1)


# plot disks
sel = np.where(tsne_results["label"] == 1)[0]
ax1.plot(tsne_results["X"][sel],tsne_results["Y"][sel],
         "o",markersize=2,markeredgecolor=def_cols[0],alpha=0.8,
         markeredgewidth=1,
         label="Disks")

# plot spheroids
sel = np.where(tsne_results["label"] == 0)[0]
ax1.plot(tsne_results["X"][sel],tsne_results["Y"][sel],
         "s",markersize=5,fillstyle='none',markeredgecolor=def_cols[1],alpha=0.8,
         markeredgewidth=1,
         label="Spheriods")

# plot stacks (use "inset_axes" to position the images in the coordinate frame of "ax1")
for ii in range(len(grid)):
    if len(STACKS[ii]) > 1:
        axins = inset_axes(ax1, width="100%", height="100%", loc='center',
                       bbox_to_anchor=(grid[ii][0]-gridsize/2,grid[ii][1]-gridsize/2,gridsize*0.8,gridsize*0.8),bbox_transform=ax1.transData)
        axins.imshow(STACKS[ii].reshape(f160w_images.shape[1],f160w_images.shape[2]),cmap=plt.get_cmap("Greys"))
    
        axins.tick_params(axis='both', which='both', bottom=False, top=False, labelbottom=False, right=False, left=False, labelleft=False)


# create the axes
ax1.set_xlabel("T-SNE dimension 1" , fontsize=14)
ax1.set_ylabel("T-SNE dimension 2",fontsize=14)
ax1.legend(loc="best",fontsize=14)
ax1.tick_params(axis='both', which='major',length=5, labelsize=12,pad=10)
ax1.tick_params(axis='both',which='minor',length=3)
ax1.minorticks_on()

#plt.savefig("tutorial1c_fig3.pdf",bbox_inches="tight")
plt.show()
```

```python
# ========================== 7. VISUALIZE RESULTS (PART 2) ==========================
# Another thing to look at is the specific star-formation rate (sSFR) of galaxies. The sSFR is related to the
# growth of a galaxy or the rate of star formation per unit mass (sSFR = SFR / M).
# Here, we use the additional info for our galaxies (specifically their RA/DEC coordinates) to match
# them to the 3D-HST catalog (Skelton et al. 2014; Brammer et al. 2015; Momcheva et al. 2015), which contains SED fits
# to all the galaxies in the GOODS-S CANDELS field.
# For more information on 3D-HST, see the data page: https://3dhst.research.yale.edu/Data.php

## a) Load the 3D-HST catalog
sedfits = Table.read("../data/goodss_3dhst_v4.1_merged.fits")

## b) Match our galaxies to the 3D-HST catalog by using their RA and DEC coordinates
## We use here a very simple matching and take the closest galaxy withing a radius of 2 arcseconds.
## This may not be ideal and you can use a more sophisticated matching for example using
## the tools from ASTROPY/SKYCOORD (https://docs.astropy.org/en/stable/coordinates/matchsep.html)
## In the following, we add the sSFR measurements (in 1/Gyr) to the "tsne_results" table for easy access.
tsne_results["lssfr"] = np.zeros(len(tsne_results))
for ii in range(f160w_info.shape[0]):
    d = 3600*np.sqrt( ( sedfits["ra"] - f160w_info[ii][1] )**2 + ( sedfits["dec"] - f160w_info[ii][2] )**2 ) # distance in arcseconds
    
    if np.nanmin(d) < 2: # closest within 2 arcseconds
        sel = np.where(d == np.nanmin(d))[0][0]
        tsne_results["lssfr"][ii] = sedfits["lssfr"][sel] + 9 # in 1/Gyr
    else:
        print("galaxy not found!")
        tsne_results["lssfr"][ii] = -99


## c) Create a figure as before but now color-code the galaxies by their log(sSFR)
        
# create a new figure
fig = plt.figure(figsize=(10,8))
ax1 = fig.add_subplot(1,1,1)

# define color maps
colmap = plt.get_cmap("Spectral")


# Plot all galaxies with color indicating their specific SFR
s1 = ax1.scatter(tsne_results["X"],tsne_results["Y"],marker="o",alpha=1,c=tsne_results["lssfr"],edgecolors="white",s=50,linewidth=0.5,
                vmin=-3,vmax=1,cmap=colmap)

# create colorbar
cbar = plt.colorbar(s1,extend="both")
cbar.set_label(r'log(sSFR / Gyr$^{-1}$)', rotation=90,fontsize=14,labelpad=0)

# create the axes
ax1.set_xlabel("T-SNE dimension 1" , fontsize=14)
ax1.set_ylabel("T-SNE dimension 2",fontsize=14)
ax1.tick_params(axis='both', which='major',length=5, labelsize=12,pad=10)
ax1.tick_params(axis='both',which='minor',length=3)
ax1.minorticks_on()

#plt.savefig("tutorial1c_fig4.pdf",bbox_inches="tight")
plt.show()
```

```python
# ===================== APPENDIX 1: HOW CAN WE MAP GALAXIES ONTO A T-SNE MAP? ====================
# Since there is no real model for the T-SNE and no real "training sample", it is not straight forward map galaxies onto the
# T-SNE space, for example to compute the accuracy. However, that we can do is to put a
# test galaxy onto the T-SNE map (Figure above) by matching the test galaxy to the "training galaxies",
# i.e., the galaxies used to create the T-SNE map. Once we have the location of the test galaxy on the T-SNE map,
# we can use a k Nearest Neighbor to estimate the probability of it belonging to a certain label.
# The matching can be done via Chi^2 statistics or more sophisticated methods.
# In the following, we sketch this method.

## Choose a test galaxy (just take the first one)
# also remove the test galaxy from the rest of the galaxies
test_galaxy = f160w_images[0].copy()
test_label = f160w_labels[0].copy()
train_galaxies = f160w_images[1:].copy()
train_labels = f160w_labels[1:].copy()
train_tsne_results = tsne_results[1:].copy()

## Match it to the galaxies via the Chi^2 method
## Using a very simple Chi2 matching to find best matching galaxy in the training sample
chi2 = []
test_galaxy_flatten = test_galaxy.flatten()
for ii in range(train_galaxies.shape[0]):
    chi2.append( np.nansum( (test_galaxy_flatten - train_galaxies[ii].flatten())**2  ) ) # simple Chi2 SUM((test-model)^2)
idx_best = np.where( np.nanmin(chi2) == chi2)[0][0]


## Let's compare the two galaxies, just for fun! - They are both Spheroids!
fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(1,2,1)
ax1.imshow(test_galaxy)
ax1.text(0.1,0.9,["Spheroid","Disk"][f160w_labels[0]],color="white",fontsize=14,va="top",ha="left",transform=ax1.transAxes)
ax1.set_title("Test Galaxy")
ax2 = fig.add_subplot(1,2,2)
ax2.imshow(train_galaxies[idx_best])
ax2.text(0.1,0.9,["Spheroid","Disk"][train_labels[idx_best]],color="white",fontsize=14,va="top",ha="left",transform=ax2.transAxes)
ax2.set_title("Most Similar Training Galaxy")
plt.show()

## Now use simple nearest neighborhood matching to get a probability for a label for our test galaxy
## There are more sophisticated ways to do this. For example, a clustering algorithm would be better because
## the T-SNE map does not have to be uniform. Clearly there are voids and higher density regions that have
## to be taken into account
dist = np.sqrt( (train_tsne_results["X"][idx_best]-train_tsne_results["X"])**2 + (train_tsne_results["Y"][idx_best]-train_tsne_results["Y"])**2 ) # use Euclidian distance
train_tsne_results["dist"] = dist # add to table
train_tsne_results.sort("dist") # sort by distance
train_tsne_results_5nn = train_tsne_results[0:6].copy() # Include the best match galaxy
print("--- 5 nearest neighbors ---")
print(train_tsne_results_5nn)

## get a weighted label (weight by distance)
## Normalize between 1 and 2 (not 0 to 1 because else weights are not finite)
## This is a very simple way to do this.
w = train_tsne_results_5nn["dist"]
w_norm = (w-np.nanmin(w))/(np.nanmax(w)-np.nanmin(w))+1
train_label_weighted = np.round( np.nansum(train_tsne_results_5nn["label"]/w_norm) / np.nansum(1/w_norm) , 3)
print("Weighted label for test galaxy: %g" % train_label_weighted)
print("True label for test galaxy: %g" % test_label)


## Visualize on the T-SNE space
fig = plt.figure(figsize=(10,8))
ax1 = fig.add_subplot(1,1,1)

## plot disks (label = 1)
sel = np.where(tsne_results["label"] == 1)[0]
ax1.plot(tsne_results["X"][sel],tsne_results["Y"][sel],
         "o",markersize=2,markeredgecolor=def_cols[0],alpha=0.8,
         markeredgewidth=1,
         label="Disks")

## plot spheroids (label = 0)
sel = np.where(tsne_results["label"] == 0)[0]
ax1.plot(tsne_results["X"][sel],tsne_results["Y"][sel],
         "s",markersize=5,fillstyle='none',markeredgecolor=def_cols[1],alpha=0.8,
         markeredgewidth=1,
         label="Spheriods")

## Mark the 5 nearest neighbors
ax1.plot(train_tsne_results_5nn["X"],train_tsne_results_5nn["Y"],
         "o",markersize=10,fillstyle='none',markeredgecolor="red",alpha=0.8,
         markeredgewidth=1,
         label="5-NN")

## Mark the best match training galaxy, which is the one with smallest distance
ax1.plot(train_tsne_results_5nn["X"][0],train_tsne_results_5nn["Y"][0],
         "x",markersize=10,fillstyle='none',markeredgecolor="red",alpha=0.8,
         markeredgewidth=1,
         label="most similar training galaxy")

## create the axes
ax1.set_xlabel("T-SNE dimension 1" , fontsize=14)
ax1.set_ylabel("T-SNE dimension 2",fontsize=14)
ax1.legend(loc="best",fontsize=12)
ax1.tick_params(axis='both', which='major',length=5, labelsize=12,pad=10)
ax1.tick_params(axis='both',which='minor',length=3)
ax1.minorticks_on()

```
