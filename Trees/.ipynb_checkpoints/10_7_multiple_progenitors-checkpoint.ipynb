```python
# Finding the number of mergers that have more than 2 progenitors

import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append('/u/e/jb2854/Mergers_Project')
from illustris_python import illustris_python as il
sys.path.append('/u/e/jb2854/Mergers_Project')
import tools

basePath = '/x/Physics/AstroPhysics/Shared-New/DATA/IllustrisTNG/TNG100-1/output'
```

```python
# Fields we need to pull the information from the tree

fields = ['SubhaloID','NextProgenitorID','MainLeafProgenitorID','FirstProgenitorID','SubhaloMassType','SnapNum', 'DescendantID', 'SubfindID']

# Starting at z = 0
groupFirstSub = il.groupcat.loadHalos(basePath,99,fields=['GroupFirstSub'])

# Trying to use subhalo flag to identify only good galaxies. BUT this will only pick good centrals in the z=0 snapshot 
subhalo_flag = il.groupcat.loadSubhalos(basePath, 99, fields = ['SubhaloFlag'])
good_cen_idx = np.intersect1d(groupFirstSub, np.where(subhalo_flag == 1)[0])

ratio = 1.0/5.0
descIDs = np.zeros([0,3]) # Storing descendant ID and snapshot number for each merger

# For the x most massive halos at z = 0
for i in range(0, 1000, 1):
    tree = il.sublink.loadTree(basePath,99,good_cen_idx[i],fields=fields) # Load tree
    mergerData = tools.mergerInfo(tree,minMassRatio=ratio) # Pull data for all mergers on a given tree
    for i in range (len(mergerData)):
        # Store descendant IDs and snapshot number
        if (mergerData[i][0] > 0.1 and mergerData[i][1] > 0.1): # Check that FP and NP are massive enough to be galaxies
            descIDs = np.vstack((descIDs, np.array([mergerData[i][4], mergerData[i][3], mergerData[i][5]])))
```

```python
numProgenitors = np.zeros([0,3])

# Iterate over all merger entries
for i in range (len(descIDs)):
    count = 0
    for j in range (len(descIDs)):
        if (descIDs[i,0] == descIDs[j,0]):
            # Count if same descendant ID is spotted
            count +=1
    # Add 1 for number of progenitors
    count +=1
    
    numProgenitors = np.vstack((numProgenitors, np.array([count, descIDs[i,1], descIDs[i,2]])))
```

```python
plt.hist(numProgenitors[:,0])
plt.xticks([2,3,4,5])
plt.xlabel("Number of progenitors")
plt.ylabel("Frequency")
plt.tight_layout()
```

```python
plt.hist(numProgenitors[:,1], bins = np.linspace(0,99,50), color = 'k', histtype = 'step')
plt.tight_layout()
plt.xlabel()
```
