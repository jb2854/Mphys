```python
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append('/u/e/jb2854/Mergers_Project')
from illustris_python import illustris_python as il
```

```python
basePath = '/x/Physics/AstroPhysics/Shared-New/DATA/IllustrisTNG/TNG100-1/output'
```

```python
# Displaying the subhalo mass against snapshot number for the 5 most massive subhalos at z=0
fields = ['SubhaloMass', 'SubfindID', 'SnapNum']
start = 0
GroupFirstSub = il.groupcat.loadHalos(basePath,99,fields = ['GroupFirstSub'])

plt.figure(figsize = (10,10))
for i in range(start, start+5):
    tree = il.sublink.loadTree(basePath, 99, GroupFirstSub[i], fields = fields, onlyMPB = True)
    plt.plot(tree['SnapNum'], tree['SubhaloMass'],'-')
    plt.yscale('log')
    plt.xlabel('Snapshot Number')
    plt.ylabel('Total Subhalo Mass [code units]')
```

```python
#Counting the number of mergers above a mass ratio of 1/5
ratio = 1/5
fields = ['SubhaloID', 'NextProgenitorID', 'MainLeafProgenitorID', 'FirstProgenitorID', 'SubhaloMassType']

for i in range(start,start+5):
    tree = il.sublink.loadTree(basePath,99,GroupFirstSub[i],fields=fields)
    numMergers = il.sublink.numMergers(tree)
    print (GroupFirstSub[i], numMergers)
```
